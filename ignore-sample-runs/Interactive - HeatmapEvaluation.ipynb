```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the data from an Excel file
df = pd.read_excel('C:\\Users\\ayesha.amjad\\Downloads\\test.xlsx')  # Update the path to your Excel file

# Check the dataframe to ensure correct data is loaded
print(df)

# Create a heatmap
plt.figure(figsize=(10, 8))  # Set the figure size
sns.heatmap(df.corr(), annot=True, fmt=".2f", cmap='coolwarm')  # 'df.corr()' computes the correlation matrix
plt.title('Heatmap of Correlation Matrix')
plt.show()
```
